**Функция *get_list_of_links* понадобится для сбора ссылок на сами ОП как на русском, так и на английском языке. Если ссылка существует, то она заносится в список ссылок, если ее нет — вместо нее список пополняется заметкой "Отсутствует".**

```python
def get_list_of_links(some_post, StPr_code_piece):
    links = []
    for link_here in some_post:
        actualink = link_here.get('href')
        if actualink == None:
            links.append("Отсутствует")
        elif StPr_code_piece in actualink:
            links.append(actualink)
    return links
```

**Строки, которые есть в функции *get_soup*, нужно было использовать несколько раз и с разными ссылками в *requests.get*, поэтому эти строки было удобно оформить в виде небольшой функции.**

```python
def get_soup(some_link):
    response = requests.get(some_link, headers={'User-Agent':user_agent})
    html_cntnt = response.text
    some_soup = BeautifulSoup(html_cntnt, 'html.parser')
    return some_soup
```

```python
def list_united(some_list):
    all_els=[]
    for lst in some_list:
        for el in lst:
            all_els.append(el)
    return all_els
```

### 4. Поиск ссылок внутри МООК
**Очевидно, что в каждой ссылке на МООК должны быть ссылки на сами курсы. В некоторых случаях на странице МООК вместо определенных ссылок на онлайн-курсы лежит какой-нибудь документ с ними. Удобно то, что любой документ (встретились docx, pdf, xlsx) на сайте тоже представляет из себя ссылку, которую можно найти.** 

```python
moocs = df["Ссылка на МООК"]
```

```python
%%time
courses_in_links = []
a = 0
for mooc in moocs[:542]:
    if mooc != "Отсутствует":
        courses_in_link = []
        mooc_soup = get_soup(mooc)
        mooc_search = mooc_soup.find_all('a')
        if mooc == "https://perm.hse.ru/ba/hist/moocs": #здесь лежит файл .docx с ссылками
            for mooc_res in mooc_search:
                m_href = mooc_res.get('href')
                if m_href != None:
                    if m_href.endswith("docx"):
                        m = "https://perm.hse.ru" + m_href
                        courses_in_link.append(m)
        elif mooc == "https://spb.hse.ru/ba/management/moocs": #здесь лежит файл .docx с ссылками
            for mooc_res in mooc_search:
                m_href = mooc_res.get('href')
                if m_href != None:
                    if m_href.endswith("docx"):
                        m = "https://spb.hse.ru" + m_href
                        courses_in_link.append(m)
        elif mooc == "https://spb.hse.ru/ba/economics/moocs": #здесь лежит файл .xlsx с ссылками
            for mooc_res in mooc_search:
                m_href = mooc_res.get('href')
                if m_href != None:
                    if m_href.endswith("xlsx"):
                        m = "https://spb.hse.ru" + m_href
                        courses_in_link.append(m)
        elif mooc == "https://perm.hse.ru/ba/economics/moocs": #здесь лежит файл .pdf с ссылками
            for mooc_res in mooc_search:
                m_href = mooc_res.get('href')
                if m_href != None:
                    if m_href.endswith("pdf"):
                        m = "https://perm.hse.ru" + m_href
                        courses_in_link.append(m)
            
        else: #в остальных случаях на страницах находятся сами ссылки
            for mooc_res in mooc_search:
                m_href = mooc_res.get('href')
                if m_href != None:
                    if "course" in m_href:
                        if "firstcourse" not in m_href:
                            if "/ba/" not in m_href:
                                courses_in_link.append(m_href)
                    elif "openedu" in m_href:
                        courses_in_link.append(m_href)
        courses_in_links.append(courses_in_link)
    else:
        courses_in_links.append("Отсутсвует")
    a+=1
```
